# Project 15: Find the Smallest Number in a List

```python
# This project teaches how to find the smallest number using loops and lists
```

```python
# Step 1: Ask the user to enter numbers separated by commas
numbers_input = input("Enter numbers separated by commas: ")
```

```python
# Step 2: Convert input string to a list of numbers
numbers_list = [float(x) for x in numbers_input.split(",")]
```

```python
# Step 3: Initialize the smallest number with the first element
smallest = numbers_list[0]
```

```python
# Step 4: Compare each number in the list
for num in numbers_list:
    if num < smallest:
        smallest = num
```

```python
# Step 5: Print the smallest number
print(f"The smallest number in the list is: {smallest}")
```
